```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import StaleElementReferenceException , NoSuchElementException
import re
```

```python
driver =webdriver.Chrome("D:\web driver\chromedriver.exe")
driver.get('https://www.flipkart.com/')
```

# Smartphones

```python
u1=[]
s1=[]
f1=[]
sh1=[]
time.sleep(2)
start=1
end=100

for page in range(start,end):
    url_1 = driver.find_elements_by_xpath('//a[@class="ge-49M _2Kfbh8"]')
    for i in url_1:
        u1.append(i.get_attribute('href'))
        
    url_2 =  driver.find_elements_by_xpath('//a[@class="ge-49M"]')
    for i in url_2:
        u1.append(i.get_attribute('href'))
time.sleep(5)

for i in u1:
    driver.get(i)
    time.sleep(2)
    
    
    try:
        j = driver.find_element_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
        s1.append(j.text)
            
    except NoSuchElementException:
        s1.append('-')
        
        
    time.sleep(1)   
    
    try:
        k = driver.find_element_by_xpath('//p[@ class="_2-N8zT"]')
        sh1.append(k.text)
            
    except NoSuchElementException:
        sh1.append('-')
        
    time.sleep(1)    
    
    try:
        l = driver.find_element_by_xpath('//div[@class="t-ZTKy"]/div/div')
        f1.append(l.text)
            
    except NoSuchElementException:
        f1.append('-')
            
    
time.sleep(2)   

# Make dataframe
df = pd.DataFrame({'Star Ratings':s1,'Short Review':sh1,'Full Review':f1})

df
```

```python
# saving the dataframe

df.to_csv('smartphone.csv')
```

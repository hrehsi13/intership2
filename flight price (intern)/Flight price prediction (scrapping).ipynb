```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import StaleElementReferenceException , NoSuchElementException
import re
from selenium.webdriver.common.by import By
import requests
```

```python
driver =webdriver.Chrome("D:\webdriver\chromedriver.exe")
driver.get('https://www.google.com/travel/flights?tfs=CBwQARoaagwIAhIIL20vMGRsdjASCjIwMjItMTEtMDUaGhIKMjAyMi0xMS0wOXIMCAISCC9tLzBkbHYwcAGCAQsI____________AUABSAGYAQE')
```

```python
air=[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@ class="Ir0Voe"]/div[2]')
    for i in na:
        air.append(i.text)

    
air     
```

```python
d_time=[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@ class="Ir0Voe"]/div[1]')
    for i in na:
        d_time.append(i.text[0:5])

    
d_time 
```

```python
a_time=[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@ class="Ir0Voe"]/div[1]')
    for i in na:
        a_time.append(i.text.replace("\n","")[10:15])

    
a_time 
```

```python
duration =[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@class="Ak5kof"]/div')
    for i in na:
        duration.append(i.text)

duration    
```

```python
t_stops =[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@class="EfT7Ae AdWm1c tPgKwe"]/span[1]')
    for i in na:
        t_stops.append(i.text)

t_stops
```

```python
route =[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//span[@class="sSHqwe tPgKwe ogfYpf"]')
    for i in na:
        route.append(i.text)

route
```

```python
price =[]
start=0
end=10
for page in range(start,end):
    na=driver.find_elements(By.XPATH,'//div[@class="U3gSDe"]/div[1]')
    for i in na:
        price.append(i.text)

price
```

```python
# Dataframe
df = pd.DataFrame({})
df['Airlines']=air
df['Departure_Time']=d_time
df['Route']=route
df['Arrival_Time']=a_time
df['Flight_Duration']=duration
df['Total_Stops']=t_stops
df['Price']=price

df
```

```python
# saving the dataframe into a csv file

df.to_csv('flight_final.csv')
```
